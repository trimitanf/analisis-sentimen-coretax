```python
!pip install deep-translator

from deep_translator import GoogleTranslator

# Fungsi untuk translate
def translate_to_english(text):
    # Jika string kosong, langsung kembalikan tanpa translate
    if not isinstance(text, str) or text.strip() == '':
        return ''  # Atau bisa juga: return 'no text'
    try:
        translated = GoogleTranslator(source='auto', target='en').translate(text)
        return translated if translated else ''
    except:
        return ''  # Jika error translasi, tetap return string kosong

# Menerapkan pada kolom tokenizing
df['translation'] = df['tokenizing'].astype(str).apply(translate_to_english)
df.head()
```
